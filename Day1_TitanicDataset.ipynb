```python
#Checking the null value using heatmap

plt.figure(figsize=(15,15))
sns.heatmap(data.isnull(),cmap='viridis',yticklabels=False)
```

```python
#Drop the Cabin as it has 687 null value out of 891

data.drop('Cabin',axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Age'].value_counts()
```

```python
#fetch the value of age between 18 and 36 as per data most of the passenger of this age so we create the list of it,
#now based on this will fill the null values
l=data[data['Age'].between(18,36)]['Age']
l.mean()
```

```python
data['Age']=data['Age'].fillna(l.mean())
```

```python
data.isnull().sum()
```

```python
#Handle the missing value of "Embarked Column"

data['Embarked'].value_counts()
```

```python
data['Embarked'].mode()[0]
```

```python
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
#All the missing values handled

#Display the How many survived and not survived
sns.countplot(data['Survived'])
```
